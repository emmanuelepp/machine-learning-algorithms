### Linear Regression

##### What is linear regression?

A linear regression is a statistical method used to predict the value of a variable based on the value of another variables. 
The idea is to draw a line in 2D or a plane in 3D that best represents the relationship between these variables. This line or plane is then used to predict new values.

Lineal regression Formula:



- **y**: The dependent variable that we are trying to predict.
- **x**: The independent variable that we use to make the prediction.
- **β₀**: The intercept, or the value of y when x = 0.
- **β₁**: The slope of the line, which indicates how much y changes with a unit change in x.
- **ϵ**: Represents the error of the model, that is, the difference between the observed values and the values that the model predicts.


#### Example



Note: <br>
The dashed line in the graph represents the distance between each data point and the regression line. This distance is known as the residual or error and shows how much the actual data value deviates from the value predicted by the regression line.

#### Implementation

```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
from sklearn.metrics import r2_score
```

```python
# Generate Data
X, y, coef = make_regression(n_samples=100, n_features=1, noise=20, coef=True, random_state=1)
```

```python
# Create and create the model
model = LinearRegression()
model.fit(X, y)
```

```python
# Evaluate
y_pred = model.predict(X)
print(f"Determination coefficient R²: {r2_score(y, y_pred)}")
```

```python
# Graph
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Real data')
plt.plot(X, y_pred, color='red', label='Regression line')
plt.title('Simple Linear Regression')
plt.xlabel('Independent variable X')
plt.ylabel('Dependent variable y')
plt.legend()
plt.show()
```
